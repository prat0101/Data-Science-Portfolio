**Task 4: Provide Statewise distribution of compliants. Find state with maximum no. of complaints and state with maximum no. of unresolved complaints**

```python
df[['State','new_status','Status']].groupby(['State','new_status']).count()
###Table for statewise status of complaints
```

```python
df['State']=df['State'].apply(lambda x: x.lower().strip())###Lower case names of all states as some values were repeating due to different case of letters
```

```python
df[['State','new_status','Status']].groupby(['State','new_status']).count().unstack()
```

```python
data=df[['State','new_status','Status']].groupby(['State','new_status']).count().unstack()
data.plot(figsize=(18,6),kind= 'bar',stacked=True,ylabel='Complaint Status')
##Stacked Bar chart for statewise distribution of total complaints, Open complaints and closed complaints
```

*From the chart it can be seen that Georgia state has maximum number of complaints* 

```python
data['total_complaints']=data.sum(axis=1)
data
###Acquire Column of Total Complaints
```

```python
data.columns
```

```python
data.columns=[i+'_'+j for i,j in data.columns]###Convert multiindex column to single index column
```

```python
data
```

```python
data['unresolved_complaints']=(data['Status_Open']/data['total_complaints_'])*100
data
###Acquire percentage of unresolved complaints
```

```python
max(data['unresolved_complaints'])
```

```python
data.loc[data['unresolved_complaints']==50]
```

*As Kansas State has only 2 total complaints. There is no meaning in calculating percentage of unresolved complaints.
We should calculate this percentage for states where there are atleast 50 complaints.*

```python
data1=data.loc[data['total_complaints_']>50]###View the states with total complaints greater than 50
data1
```

```python
data1.loc[data1['unresolved_complaints'].idxmax()].to_frame() 
####State corresponding to maximum unresolved complaint provided that it has total complaints greater than 50
```

*Tennessee State has maximum percentage of unresolved complaints. This is meaningful as it has significant number of total complaints.*
